```python
import numpy as np
import pandas as  pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Read file and display it. This is to get an idea of the data I am working with

```python
myDataFromFile = pd.read_csv("datasets/007ExerciseFile.csv")
display(myDataFromFile)
```

Make an informative histogram of the mutation rates of the 4 strains

```python
sns.displot(myDataFromFile, x="Mutation_rate", hue="Strain", kind="hist", alpha=0.2, bins=20)
```

Make an informative KDE plot

```python
sns.displot(myDataFromFile, x="Mutation_rate", hue="Strain", kind="kde", fill=True, alpha=0.4, bw_adjust=0.8)
```

Make a violin plot of the data

```python
sns.violinplot(data=myDataFromFile, x="Mutation_rate", y="Strain")
```

Make a boxplot of the data

```python
sns.boxplot(data=myDataFromFile, x="Mutation_rate", y="Strain")
```

Make a boxplot with an overlaid strip chart

```python
sns.boxplot(data=myDataFromFile, x="Mutation_rate", y="Strain")
sns.stripplot(data=myDataFromFile, x="Mutation_rate", y="Strain", alpha=0.4)
```

Selected graph with English labels

```python
sns.boxplot(data=myDataFromFile, x="Mutation_rate", y="Strain")
sns.stripplot(data=myDataFromFile, x="Mutation_rate", y="Strain", alpha=0.4)
plt.title("Mutation Rate of 4 Different Strains")
plt.xlabel("Mutation Rate")
plt.ylabel("Strain Type")
plt.yticks([0, 1, 2, 3], ['Alpha', 'Beta', 'Gamma', 'Delta'])
```

The graph demonstrates that each strain is mutating at a different rate with the Alpha strain mutating the fastest and the Gamma strain mutating the slowest.
