```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mdff = pd.read_csv("datasets/008ExerciseFile.csv")
display(mdff)
```

Try a straight line through the data

```python
mdff.plot(x='rel_time', y='m_level', kind='scatter', color='r')
```

```python
sns.set_theme()
sns.relplot(data=mdff, x='rel_time', y='m_level')
```

```python
sns.lmplot(data=mdff, x='rel_time', y='m_level')
```

The straight line does not depict a good line of fit for the data. Lets check the residual graph

```python
sns.residplot(data=mdff, x='rel_time', y='m_level')
```

The graph shows a pattern to the residuals. Because the residuals are not random, we can reject that the straight line is a fit for the data

Now try to find a fit that works

```python
sns.lmplot(data=mdff, x='rel_time', y='m_level', order=2)
```

```python
sns.lmplot(data=mdff, x='rel_time', y='m_level', order=3)
```

```python
sns.lmplot(data=mdff, x='rel_time', y='m_level', order=4)
```

The line at order 3 and 4 appear to fit the data better. Let's check the residual graphs for these orders

```python
g = sns.residplot(data=mdff, x='rel_time', y='m_level', order=3)
```

```python
h = sns.residplot(data=mdff, x='rel_time', y='m_level', order=4)
```

Both graphs show a good line of fit, but order=3 is a simpler model so it is the one we will go with

```python
myfit = np.polyfit(x=mdff['rel_time'], y=mdff['m_level'], deg=3)
print(myfit)
```

```python
fitvals = np.polyval(myfit, mdff['rel_time'])
myres = mdff['m_level'] - fitvals
sns.displot(myres)
```

Our residuals shows a relatively normally distributed graph

```python
np.std(myres)
```

Let's make our graph pretty and combine the information we know

```python
fig = sns.lmplot(data=mdff, x='rel_time', y='m_level', scatter_kws={'color': 'teal'}, line_kws={'color': 'black'}, order=3)
plt.title("Hormone Fluctuations Over a Single Day")
plt.xlabel("Relative Time of Day")
plt.ylabel("Relative Melatonin Level")
sns.set_style("white")
```

The line of fit in our data is best captured by a third order polynomial. Our findings show the equation to be: y = 107.8 + -50.99x + 1.54x^2 + 0.52x^3 + N(0,46)
