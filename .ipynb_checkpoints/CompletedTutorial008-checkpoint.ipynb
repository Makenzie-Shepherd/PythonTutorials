Plotting Relational Data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mdff = pd.read_csv("datasets/008TutorialDataFile1.csv")
display(mdff)
```

```python
mdff.plot(x='X', y='Y', kind='scatter', color='r')
```

Plot shows a relationship

y = a + bx + err

a : y-intercept; b : slope; err : error/variability

```python
sns.set_theme() #set theme to seaborn's default
sns.relplot(data=mdff, x='X', y='Y')
```

```python
sns.lmplot(data=mdff, x='X', y='Y') #plots a fitted line to the data aka a 'linear model'
```

Takeaways: there is a relationship between x and y & the random variability limits our ability to predict any one particular y value to a given x value

```python
myfit = np.polyfit(x=mdff['X'], y=mdff['Y'], deg=1) #estimates the slope and y-intercept
print(myfit)
```

```python
sns.residplot(data=mdff, x='X', y='Y') #plots the residuals
```

```python
fitvals = np.polyval(myfit, mdff['X'])
```

```python
myres = mdff['Y'] - fitvals
```

```python
sns.displot(myres, kind='kde')
```

```python
np.std(myres) #standard deviation
```

Line of Fit: y = 4.5x + 242 + N(0, 36)

Third Data Set

```python
mdff = pd.read_csv("datasets/008TutorialDataFile3.csv")
display(mdff)
```

```python
sns.relplot(data=mdff, x='X', y='Y')
```

```python
sns.lmplot(data=mdff, x='X', y='Y')
```

```python
sns.residplot(data=mdff, x='X', y='Y')
```

Graphs shows a pattern to the residuals. The reason it the model is a second order polynomial: y = a + bx + cx^2 + N(0,1)

```python
sns.lmplot(data=mdff, x='X', y='Y', order=2)
```

```python
g = sns.residplot(data=mdff, x='X', y='Y', order=2)
```

```python
myfit = np.polyfit(x=mdff['X'], y=mdff['Y'], deg=2)
print(myfit)
```

```python
fitvals = np.polyval(myfit, mdff['X'])
myres = mdff['Y'] - fitvals
sns.displot(myres)
```

```python
np.std(myres)
```

y = 7.3 + 0.02x + 0.7x^2 + N(0,98)
