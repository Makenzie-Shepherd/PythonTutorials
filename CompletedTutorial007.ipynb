```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Read file and display it

```python
myDataFromFile = pd.read_csv("datasets/007DataFile.csv")
display(myDataFromFile)
```

Tidy data: includes a grouping variable that makes it easier to segregate the data

Visualization - Histogram. 
Argument 1: data frame; Argument 2: value variable assigned to the x axis; Argument 3: variable color of bars

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="hist", alpha=0.3)
```

Visualization: Kernel Density Estimate (KDE) plots.
Represents the smoothness of continuous data

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="kde")
```

Alpha is the level of transparency

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="kde", fill=True, alpha=0.2)
```

Combine both types of graphs

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="hist", kde=True)
```

Too many bins

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="hist", bins=100)
```

Too few bins

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="hist", bins=3)
```

Binwidth adjusts width based on the value on the x-axis

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="hist", binwidth=.5)
```

Bw_adjust argument adjusts the bandwidth of the kernel.
The larger the bandwidth, the smoother the plot will appear

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="kde", bw_adjust=3)
```

The smaller the bandwidth, the more wiggly the plot will appear

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="kde", bw_adjust=0.5)
```

Visualization: Empirical Cumulative Density Estimate plot.
Better for revealing small but systematic shifts in the mean

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="ecdf")
```

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="kde", cumulative=True)
```

Add grid lines to the graph

```python
sns.set_style("darkgrid")
sns.displot(myDataFromFile, x="Value", hue="Group", kind="ecdf")
```

```python
sns.displot(myDataFromFile, x="Value", hue="Group", kind="ecdf")
myTickMarks = np.arange(0, 1, 0.1)
plt.yticks(myTickMarks)
```

Visualization: Categorical or Strip Plots. 
The categorical data on the x-axis and numerical data on the y-axis

```python
sns.catplot(data=myDataFromFile, x="Group", y="Value")
```

```python
sns.catplot(data=myDataFromFile, x="Group", y="Value", alpha=0.5)
```

Visualization: Violin plots. KDE plot flipped on its side, separated along the x axis, and plotted along with a mirror image

```python
sns.violinplot(data=myDataFromFile, x="Group", y="Value")
```

```python
sns.violinplot(data=myDataFromFile, x="Group", y="Value", inner="stick")
```

Visualization: Boxplots. Displays 5 summary numbers in graph form

```python
sns.boxplot(data=myDataFromFile, x="Group", y="Value")
```

```python
sns.boxplot(data=myDataFromFile, x="Group", y="Value")
sns.stripplot(data=myDataFromFile, x="Group", y="Value", alpha=0.42)
```
